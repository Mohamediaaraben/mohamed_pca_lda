# Mouse Click 1

```python
read_path = "../../data/images/"
write_path = "../../data/relabelled/"
```

```python
def onclick(event):
    global img_path
    global img
    
    #Get x,y of click
    x = round(event.xdata)
    y = round(event.ydata)
    
    #Save image with new x,y
    path = img_path[0]
    name = path.split("/")[-1].split("_")[-1]
    new_name = "{}_{}_{}".format(x,y,name)
    
    mpimg.imsave(write_path + new_name, img)
    
    # Clear plot
    plt.clf()
    
    if len(img_path) > 0:
        
         #Remove first instance of img_path 
        img_path = img_path[1:]
        
        #Display next image
        path = img_path[0]
        img =mpimg.imread(path)

        plt.imshow(img)
        plt.show()
```

```python
%matplotlib tk
global img_path
global img

#Get all image paths
img_path = glob.glob(read_path + "/*.jpg")

#Load first image
path = img_path[0]
img = mpimg.imread(path)

#Add an interactive widget to figure 
fig = plt.figure(figsize=(5,5))
cid = fig.canvas.mpl_connect('button_press_event', onclick)

plt.imshow(img)
plt.show()
```

# Mouse Click 2

```python
read_path = "../../data/images/"
write_path = "../../data/relabelled/"
```

```python
def onclick(event):
    global img_path
    
    #Get x,y of click
    x = round(event.xdata)
    y = round(event.ydata)
    
    #Image that will be saved
    save_img = mpimg.imread(img_path[0])
    
    #Save image with new x,y
    path = img_path[0]
    name = path.split("/")[-1].split("_")[-1]
    new_name = "{}_{}_{}".format(x,y,name)
    mpimg.imsave(write_path + new_name, save_img)
    
    # Clear plot
    plt.clf()
    
    #Load next image
    if len(img_path) > 0:
        
        img_path = img_path[1:]
        
        #Get current x,y
        path = img_path[0]
        name = path.split("/")[-1]
        x = int(name.split("_")[0])
        y = int(name.split("_")[1])

        #Load image and draw circle 
        img = mpimg.imread(path)
        img = cv2.circle(img, (x, y), 8, (0, 255, 0), 3)

        plt.imshow(img)
        plt.show()
```

```python
%matplotlib notebook
global img_path

#Get all image paths
img_path = glob.glob(read_path + "/*.jpg")

path = img_path[0]
name = path.split("/")[-1]
x = int(name.split("_")[0])
y = int(name.split("_")[1])

#Load first image and add circle
img = mpimg.imread(path)
img = cv2.circle(img, (x, y), 8, (0, 255, 0), 3)

#Add an interactive widget to figure 
fig = plt.figure(figsize=(5,5))
cid = fig.canvas.mpl_connect('button_press_event', onclick)

plt.imshow(img)
plt.show()
```

# Keyboard press 

```python
read_path = "../../data/images/"
write_path = "../../data/relabelled/"
```

```python
def onpress(event):
    global img_path
    global IDs
    global labels
    
    #Get key 
    key = event.key
    
    #Get image name
    path = img_path[0]
    ID = path.split("/")[-1]
    
    
    if key not in ["l","r","d"]:
        print("Invalid Key")
    else: 
        
        if key == "l":
            IDs.append(ID)
            labels.append("left")
            
        elif key == "r":
            IDs.append(ID)
            labels.append("right")
        
        # Clear plot 
        plt.clf()

        if len(img_path) > 0: 
            
            img_path = img_path[1:]
            
            #Display next image with count 
            path = img_path[0]
            img = mpimg.imread(path)
            plt.text(0, 15, len(img_path),color='r',size=20)
            
            plt.imshow(img)
            plt.show()
            
```

```python
%matplotlib notebook
global img_path 
global IDs 
global labels
IDs = []
labels = []

#Get all image paths
img_path = glob.glob(read_path + "/*.jpg")


fig = plt.figure(figsize=(5,5))

#Load first image and add count
path = img_path[0]
img = mpimg.imread(path)
plt.text(0, 15, len(img_path),color='r',size=20)

#Add an interactive widget to figure 
cid = fig.canvas.mpl_connect('key_press_event', onpress)

plt.imshow(img)
plt.show()
```
